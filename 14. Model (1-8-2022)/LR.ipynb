```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x=np.array([13,51,20,72,91,67]).reshape((-1,1))
y=np.array([34,82,25,48,64,11])
print(x)
print(y)
print()
model=LinearRegression().fit(x,y)
print("x,y score : ",model.score(x,y))
print(model.intercept_)
print(model.coef_)
print()
ypred=model.predict(x)

print("x,Ypred score : ",model.score(x,ypred))
plt.scatter(x,y,c='black')
plt.plot(x,ypred,color='green')
plt.show()
```
